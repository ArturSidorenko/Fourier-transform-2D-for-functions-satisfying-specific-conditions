## Теория. Одномерный ряд Фурье 

Ряды Фурье - очень часто используемая конструкция, особенно при решении уравнений в частных производных Рассмотрим конструкцию одномерного ряда Фурье. Пусть дано гильбертово пространство $L^2$ со скалярным произведением $(a, b)$. Рассмотрим в нём ортонормированную систему (ОНС) ${\{e_i\}}_{i = 1}^{\infty}$, т.е. систему элементов, такую что $\forall i, j: (e_i, e_j) = \delta^i_j$. 

Система называется *полной*, если $\forall x \in L^2: (x = 0 \iff \forall i: (x, e_i) = 0)$, т.е. нет ненулевого элемента, ортогонального всем векторам ОНС сразу. 

ОНС называется *базисом*, если для всех $x \in L^2$, если сумма его ряда Фурье \begin{gather*} \sum_{i = 1}^{\infty} c_ie_i = x \end{gather*}\ , где $c_i = (x, e_i)$, а предел понимается в среднеквадратическом смысле, т.е. по норме в $L^2$.

 ОНС называется *замкнутой*, если верно равенство Парсеваля \begin{gather*}\sum_{i = 1}^{\infty} |c_i|^2 = ||x||^2 \end{gather*}. 

__Теорема__. Пусть дано сепарабельное гильбертово пространство $L^2$. Пусть дана ОНС ${\{e_i\}}_{i = 1}^{\infty}$. Тогда следующие условия эквивалентны:
1. ОНС является базисом
2. ОНС полна
3. ОНС замкнута
4. Конечные линейные комбинации элементов ОНС плотны в $L^2$, т.е. $Cl\{c_1e_1 + \dots + c_ke_k \,| \,k \in \mathbb{N}, \, c_i \in \mathbb{C} \}= L^2$

На курсах математического и функционального анализа доказывается, что данную теорему можно применить к классическому случаю, когда дано пространство измеримых (комплекснозначных) на $[-\pi, \pi]$ функций, квадрат которых интергируем, а в качестве ОНС выбираются функции  $\exp(ikx)$, где $k \in \mathbb{Z}$ с соответствующей нормировкой . Другими словами, пространство $L^2[-\pi, \pi]$ оказывается сепарабельным, а система $\frac{1}{\sqrt{2\pi}}\exp(ikx)$, где $k \in \mathbb{Z} \setminus \{0\}$, и $\frac{1}{\sqrt{2\pi}}$, будет базисом в этом пространстве. 

Данная теорема, несмотря на всю общность, гарантирует только сходимость ряда Фурье в среднеквадратическом. Это не даёт поточечную сходимость. Поточечную сходимость можно получить при помощи признака Дини.

__Определение__. Функция   $\,f : U \rightarrow \mathbb{C}$ , заданная в некой проколотой окрестности точки $x \in \mathbb{R}$, удовлетворяет в этой точке условиям Дини, если:
1. существуют оба предела $f(x_{-}) = \lim_{t \rightarrow +0} f(x-t) \,$ и   $f(x_{+}) = \lim_{t \rightarrow +0} f(x+t) \,$;
2. Интеграл \begin{gather*}\int_{0}^{\epsilon} \frac{ ( f(x-t) - f(x_{-}) ) + ( f(x+t) - f(x_{-}) ) }{t} dt \end{gather*}  сходится абсолютно при нектором положительном $\epsilon$.

__Теорема (Признак Дини)__. Пусть функция  $ \, f : \mathbb{R} \rightarrow \mathbb{C} $ является $2\pi$ периодической и абсолютно интегрируемой (по Риману) на $[-\pi, \pi]$. Пусть она удовлетворяет условиям Дини в точке $x \in \mathbb{R}$. Тогда её ряд Фурье сходится в этой точке, и  \begin{gather*}\sum_{-\infty}^{\infty} c_k \exp(ikx) = \frac{f(x_{-}) + f(x_{+})}{2} \end{gather*}. 

Очевидно, что гёльдеровы и липшицевы функции удовлетворяют этому условию. В частности, непрерывно-дифференцируемые функции на отрезке имеют сходящийся к ним же ряд Фурье. Для гладких функций есть оценки скорости сходимости ряда Фурье к функции. 

__Теорема__. Если функция $f: [-\pi, \pi] \rightarrow \mathbb{C}$ такова, что $f \in C^{(m-1)} [-\pi, \pi]$,  $m \in \mathbb{N} \,$,  $f^{(j)} (-\pi) = f^{(j)} (\pi)$, $j = 0, \dots, m-1$, и $f$ имеет кусочно-неперерывную производную f^(m), то тогда её ряд Фурье сходится абсолютно и равномерно к функции, и причём отклонение n-ой частичной суммы $S_n(x)$ ряда от $f(x)$ на всём отрезке $[-\pi, \pi]$ иммет оценку $\begin{gather*} ||f - S_n||_{C[-\pi, \pi]} = o(n^{\frac{1}{2}-m}) \end{gather*}$.

# Численное преобразование Фурье в одномерном случае

## Случай 1

Дана функция $f: [0,1] \rightarrow \mathbb{R}$. Известно, что она удовлетворяет граничным условиям $f(0) = 0$, $f'(1)=0$. Требуется найти дискретное преобразование Фурье. Для этого предлагается разделить отрезок на промежутки длины $h$ в количестве $N$ так, чтобы $(1 + h/2) = Nh$, т.е. разбиение немоного уходит вправо. Получим, что $h = \frac{2}{2N-1}$. Даны числа $f_k = f(kh), \, k = 0,\dots, N-1$. Предлагается на правом конце проводить чётное продолжление, т.е. положить $f_N = f_{N-1}$.

Вначале докажем, что функцию с данными граничными условиями можно разложить в ряд Фурье вида \begin{gather*} \sum_{i=0}^{\infty} c_k \sin (\pi(k + \frac{1}{2})x) \end{gather*} Продолжим нашу функцию вправо чётным образом, а влево - нечётным образом. Получим функцию $f$ на отрезке $[-2, 2]$. Она разложится в ряд Фурье вида \begin{gather*} \frac{a_0}{2} + \sum_{k = 1}^{\infty} b_k \sin(\frac{\pi}{2} kx) + a_k \cos(\frac{\pi}{2}  kx) \end{gather*} Заметим теперь, что т.к. $f$ нечётна, то коэффициенты при косинусах пропадут. Далее рассмотрим $f$ на отрезке $[1, 2]$. Если взять функцию $\sin(\frac{\pi}{2} kx)$ с чётным $k$, выписать интеграл для $b_k$: \begin{gather*} \int_0^2 f(x) sin (\frac{\pi}{2} kx) dx = -\int_0^2 f(2-t) (sin (\frac{\pi}{2} kt)) dt = -\int_0^2 f(t) (sin (\frac{\pi}{2} kt)) dt \end{gather*} где $t = 2 -x $, то получим, что этот интеграл равен минус себе, т.е. нулю. Остаются только синусы с нечётными параметрами, т.е. ряд получит вид  \begin{gather*}  \sum_{k = 0}^{\infty} b_k \sin(\pi (k + \frac{1}{2}) x) \end{gather*}

Теперь попробуем сделать дискретизацию задачи. У нас размерность пространства чисел $f_k$ равна $N-1$. Тогда потребуется $N-1$ линейно независимых векторов вида $ \sin(\pi (l + \frac{1}{2})kh ) $. Покажем, что они образуют ортогональный базис, если ввести скалярное произведение \begin{gather*} <f, g> = \sum_{k=0}^{N-1} f_k g_kh   \end{gather*}.

Рассмотрим лучше отрезок $[0, 2]$. В него укладывается ровно $2N-1$ отрезочков длины $h$. Учтём симметрию относительно середины отрезка, примем $M = 2N-1$ и получим, что для $l \neq m$

\begin{gather*}2<\sin(\pi (l + \frac{1}{2})kh ), \sin(\pi (m + \frac{1}{2})kh )> = \\ h \sum_{k=0}^{M} \sin(\pi (l + \frac{1}{2})kh ) \sin(\pi (m + \frac{1}{2})kh ) = \frac{h}{2} \sum_{k=0}^{M} - \cos(\pi (l + m + 1) kh) + \cos(\pi (l-m) kh)\end{gather*}

Вспомним, что $\sum_{k=1}^{M} cos(ky) = \cos(\frac{M+1}{2} y) \frac{\sin\frac{1}{2}My}{\sin\frac{1}{2} y}$

Тогда  \begin{gather*} -4<\sin(\pi (l + \frac{1}{2})kh ), \sin(\pi (m + \frac{1}{2})kh )> = \\
h\cos( (\pi (l + m + 1) \frac{2N}{2N-1} ) \frac{\sin((\pi (l + m + 1))}{\sin(0.5(\pi (l + m + 1) h)} - \\
h\cos(\frac{N}{2} \pi (l-m) h) \frac{\sin(\pi (l-m))}{\sin(0.5\pi (l-m) h)} = 0
 \end{gather*}

Для $l = m$ применим аналогичный приём:

\begin{gather*} 2<\sin(\pi (m + \frac{1}{2})kh ), \sin(\pi (m + \frac{1}{2})kh )> = \\ h \sum_{k=0}^{M} \sin^2(\pi (m + \frac{1}{2})kh ) = \frac{1}{2} \sum_{k=0}^{M} 1 - \cos(\pi 2m kh) = \\ \frac{Mh}{2} - h\sin( (\pi (2m + 1) \frac{2N}{2N-1} ) \frac{\sin((\pi (2m + 1))}{\sin(0.5(\pi (2m + 1) h)} = \frac{Mh}{2} = 1 \end{gather*}

Отсюда получим, что нормы синусов равны $\frac{1}{\sqrt{2}}$.

Для N промежутков нужен N-1 базисный элемент.

Подведём итог в иде теоремы.

__Теорема__. Пусть дано пространство $\mathbb{R}^{n+1}$ с элементами $(x_0, x_1, \dots, x_n)$ и описанным выше скалярным произведением. Рассмотрим гиперплоскость $x_0 = 0, \, x_{n-1} = x_{n}$. Тогда система элементов $\{(\sqrt{2}\sin(\pi(k+\frac{1}{2})ih), \, i = 0\dots n), \, k=0\dots n-2\}$ будет ортобазисом данной гиперплоскости.

## Случай 2 

Дана функция $f: [0,1] \rightarrow \mathbb{R}$. Известно, что она удовлетворяет граничным условиям $f'(0) = 0$, $f(1)=0$. Требуется найти дискретное преобразование Фурье. Для этого предлагается разделить отрезок на промежутки длины $h$ в количестве $N$ так, чтобы $1 = Nh$. Даны числа $f_k = f(kh), \, k = 0,\dots, N$. 

Пользуясь аналогичным приёмом можно доказать, что данную функцию надо раскладывать в ряд по косинусам: \begin{gather*} \sum_{i=0}^{\infty} c_k \cos (\pi(k + \frac{1}{2})x) \end{gather*}.

Далее покажем, что если взять формулу для скалярного произведения \begin{gather*} <f, g> = \sum_{k=1}^{N-1} f_k g_kh + \frac{h}{2} (f_0 g_0 + f_N g_N)  \end{gather*}, то функции $\sqrt{2} \cos (\pi(k + \frac{1}{2})x)$ образуют ортобазис. Для иэтого использкем чётное продолжение разбиения на отрезок $[-1, 0]$. Получим тогда, что \begin{gather*}2<\cos(\pi (l + \frac{1}{2})kh ), \cos(\pi (m + \frac{1}{2})kh )> = h \sum_{k=-N+1}^{N} \cos(\pi (l + \frac{1}{2})kh ) \cos(\pi (m + \frac{1}{2})kh ) = \\ \frac{h}{2} \sum_{k=-N+1}^{N}  \cos(\pi (l + m + 1) kh) + \cos(\pi (l-m) kh)\end{gather*}. Хотелось бы поменять индекс суммировния. Выберем $j = k + N $. Тогда для целых $m$ $\cos(\pi m kh) = \cos(\pi m (j - N)h) = \cos(\pi m jh - hN\pi m) = (-1)^m \cos(\pi m jh)$. Пользуясь этой выкладкой, поменяем индексацию: \begin{gather*}\frac{h}{2} \sum_{k=-N+1}^{N}  \cos(\pi (l + m + 1) kh) + \cos(\pi (l-m) kh) =\\ \frac{h}{2}  \sum_{k=1}^{2N}  (-1)^{l+m+1}\cos(\pi (l + m + 1) kh) + (-1)^{l-m}\cos(\pi (l-m) kh)\end{gather*}.

Подсчитаем отдельно для натуральных $m$ сумму, используя тригонометрические формулы: \begin{gather*}\sum_{k=1}^{2N}  (-1)^{m}\cos(\pi m kh) = (-1)^{m}  \cos(\frac{2N+1}{2N}) \frac{\
\sin(\pi m )}{\sin(0.5\pi mh)} = 0 \end{gather*}

Тогда получим, что $2<\cos(\pi (l + \frac{1}{2})kh ), \cos(\pi (m + \frac{1}{2})kh )>$ равен нулю при $l \neq m$ и единице при $l=m$. Отсюда следует доказываемое утверждение о том, что функции $\sqrt{2} \cos (\pi(k + \frac{1}{2})x)$ образуют ортобазис. Для N промежутков требуется N базисных эл-ов.

Подведём итог в иде теоремы.

__Теорема__. Пусть дано пространство $\mathbb{R}^{n+1}$ с элементами $(x_0, x_1, \dots, x_n)$ и описанным выше скалярным произведением. Рассмотрим гиперплоскость $x_n = 0$. Тогда система элементов $\{(\sqrt{2}\cos(\pi(k+\frac{1}{2})ih), \, i = 0\dots n), \, k=0\dots n-1\}$ будет ортобазисом данной гиперплоскости.

# Двумерное преобразование Фурье

Для расчёта двумерного преобразования Фурье при заданных краевых условиях нужно отдельно исследовать условия для оси x, отдельно для оси y, в качестве сетки для плоскости взять прямое произведение сеток на осях. Скалярное произведение вырабатывается, опять же, исходя из того, что происходит на осях координат отдельно. С базисом то же самое. 

В моём случае базисными функциями будут $2  \sin (\pi(k + \frac{1}{2})x) \cos (\pi(l + \frac{1}{2})y)$, а скалярное произведение функции $f$ на базисные считается по формуле \begin{gather*} <f, g> = \sum_{k=0}^{N-1} \psi_k g^{sin}_kh_x  \end{gather*} где \begin{gather*} \psi_k = \sum_{l=1}^{N-1} f_{kl} g^{cos}_lh_y + \frac{h_y}{2} (f_{0l} g^{cos}_0 + f_{Nl} g^{cos}_N)   \end{gather*}

 Другими словами, мы вначале фиксируем номер базисного элемента $(i, j)$. Вычисляем значения синусов и косинусов с соответствующими номерами. Далее высчитываем числа $\psi_k$, а потом уже - скалярное произведение. Получается, что рассчёты занимают время $O(n^3)$. Дополнительное ускорение может быть получено за счёт распараллеливания вычислений, например, при помощи библиотеки OpenMP, позволяющей в два касания ускорять работу в циклах.

Обратное преобразование, очевидно, можно провести по формуле перехода от одного базиса к другому. Если её правильно выписать, то можно добиться ускорения вычислений. Именно, пусть даны $c_{ij}$, $i, j = 1...n-1$. Тогда можем восстановить значения функции в узловых точках $(x_k, y_l)$ по формуле \begin{gather*} f(x_k, y_l) = \sum_{i=1}^{N-1} \beta_{il} sin((i-0.5)x_k) \end{gather*} где \begin{gather*}\beta_{il} = \sum_{j=1}^{N-1} c_{ij} cos((j-0.5)y_l)\end{gather*} Нам необязательно считать каждую точку по отдельности. Готовим массив для ответа, заполняя его нулями. Мы можем фиксировать $i, l$, затем сосчитать одну $\beta_{il}$, затем пройтись по всем $k$. Такая хитрая перестановка слагаемых позволяет избежать лишних пересчётов и съэкономить время расчётов.

Проверка того, что система $2  \sin (\pi(k + \frac{1}{2})x) \cos (\pi(l + \frac{1}{2})y)$  -  ортобазис, не вызывает трудностей. Для этого надо двойную сумму, представляющую скалярное произведение, переписать в виде повторной. Эта повторная сумма сама собой расщепляется в произведение сумм одинарных (каждая сумма соответсвует "своей" координате"), так что задача сводится к одномерной. То, что данную систему элементов не нужно никак дполнять, лекго видеть из соображений размерности.

## Тест на корректность преобразования

Здесь был численно проверен факт точного восстановления функции по Фурье-преобразованию на узлах решётки. Было взято несколько значений n, и затем было рассчитано максимум модуля отклонения оригинальной функции от восстановленной по коэффициентам Фурье.  Для исследований была взята функция $f(x, y) = x(y-1)e^{-x+y}.$ Легко проерить, что функция удовлетворяет граничным условиям.

```python
table = [
[      2, 5.55112e-17],
[      4, 2.77556e-16],
[      8, 3.33067e-16],
[     16, 9.99201e-16],
[     32, 1.94289e-15],
[     64, 7.04992e-15],
[    128, 1.18794e-14],
[    256, 3.35842e-14],
[    512, 6.27276e-14],
]
```

```python
df = pd.DataFrame(table)
df.columns = ['n', 'err']
```

```python
df
```

Ниже приведён график зависимости в двойной логарифмической шкале. Так как на ней мы видим линейный рост, исходные величины подчинены степенной зависимостью, причём угловой коэффициент предполагаемой прямой будет показателем степени зависимости. Ошибка растёт с измельчением в связи с накоплением погрешностей. Также была проведена прямая, соответствующая линейной регресии. Она и позволяет оценить угловой коэффициент.

```python
T = np.array(table)
X = np.log(T[:,0]) / np.log(2)
Y = np.log(T[:, 1]) / np.log(2)
```

```python
import sklearn.linear_model as lm

mod = lm.LinearRegression()
mod.fit(X.reshape(-1,1), Y)
Y_line = mod.predict(X.reshape(-1,1))
```

```python
import plotly.graph_objects as go
import plotly.io as pio
from IPython.display import Image
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=X, y=Y,
                         mode='lines+markers',
                         name='plot',
                         marker_size=18,
                        ))
fig.add_trace(go.Scatter(x=X, y=Y_line,
                         mode='lines',
                         name='linear approximation',
                          line=dict(color='firebrick', width=2, dash='dash')
                        ))

fig.update_layout(
    xaxis=go.layout.XAxis(
        title=go.layout.xaxis.Title(
            text="Разбиение, двоичный логарифм",
            font=dict(
                family="Courier New, monospace",
                size=16,
                color="#0f0f0f"
            )
        )
    ),
     yaxis=go.layout.YAxis(
        title=go.layout.yaxis.Title(
            text="Максимальное отклонение, двоичный логарифм",
            font=dict(
                family="Courier New, monospace",
                size=16,
               
                color="#0f0f0f"
            )
        )
    )
)


print('Оценка степени = ', mod.coef_[0])
```

```python
Image(pio.to_image(fig, format='png', scale=1.3))
```

## Тест на сходимость

В итоге мы получили, что зависимость от мелкости разбиения обратно квадратичная. Чтобы это проиллюстрировать, приведу ещё один график.

```python
T = np.array(table)
X = T[:,0] 
Y = T[:, 1]
Y = Y * X * X

fig = go.Figure()

fig.add_trace(go.Scatter(x=X, y=Y,
                         mode='lines+markers',
                         name='plot',
                         marker_size=18,
                        ))


fig.update_layout(
    xaxis=go.layout.XAxis(
        title=go.layout.xaxis.Title(
            text="Разбиение",
            font=dict(
                family="Courier New, monospace",
               
                color="#0f0f0f"
            )
        )
    ),
     yaxis=go.layout.YAxis(
        title=go.layout.yaxis.Title(
            text="Максимальное отклонение * Разбиение",
            font=dict(
                family="Courier New, monospace",
               
                color="#0f0f0f"
            )
        )
    )
)

Image(pio.to_image(fig, format='png', scale=1.2))
```

Здесь мы видим выход графика на прямую, что подтверждает нашу догадку.

Данная ситуация, должно быть, противоречит теореме о равномерной сходимости ряда Фурье. В силу гладкости функции убывание должно происходить быстрее любой степени. Тем не менее, тригонометрическая интерполяция - это совсем не то же самое, что и ряд Фурье, так как мы в расчётах подменили интегралы на специально подогнанные суммы, так что коэффициенты тригонометрических многочленов меняются с изменением $n$. <b> Вывод: дискретное преобразование Фурье - не лучший инструмент тригонометрической интерполяции. </b> Оно используется в других целях, таких как сжатие информации и устранение шумов.
